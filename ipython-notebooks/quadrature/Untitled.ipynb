```python
from equadratures import *
import numpy as np
```

```python
def fun(x):
    return np.exp(3*x[0] + x[1])
```

```python
x = Parameter(distribution='uniform', lower=-1., upper=1., order=4)
sparse = Basis('Sparse grid', level=20, growth_rule='linear')
Pol = Polyint([x,x], sparse)
Pol.computeCoefficients(fun)
```

```python
coeffplot2D(Pol.coefficients, Pol.multi_index, '$i_1$', '$i_2$')
```

```python
x, y, z, max_order = twoDgrid(Pol.coefficients, Pol.multi_index)
G = np.log10(np.abs(z))
Zm = np.ma.masked_where(np.isnan(G),G)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.scatter(x, y, s=100, marker='o', c=G, cmap='jet', alpha=0.3)
plt.xlim(-0.5, max_order)
plt.ylim(-0.5, max_order)
adjust_spines(ax, ['left', 'bottom'])
ax.set_axisbelow(True)
plt.show()
```
