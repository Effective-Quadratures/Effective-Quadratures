<h1>Plotting PDFs of input parameters

In this simple notebook, we demonstrate some functionality to plot PDFs of different distributions and shape parameters. For this we will only require the Parameter class in the effective quadrature suite. 

```python
%matplotlib inline
from effective_quadratures.parameter import Parameter
import numpy as np
```

<h2> Plot a Gaussian with mean 15 and a variance of 2.5

```python
X = Parameter(points=5, shape_parameter_A=15, shape_parameter_B=2.5, param_type='Gaussian')
PDF = X.getSamples(10000, graph=1, filename='parameter.eps')
```

The input to getSamples is the number of samples (which in this case was 10000) and a boolean for whether the user wants a graphical output or not. The analytical version of this plot can also be obtained via

```python
PDF = X.getPDF(300, graph=1)
```

<h2> Plot a Beta distribution with $\alpha=1.5$ and $\beta=2.5$

```python
X = Parameter(points=5, shape_parameter_A=1.5, shape_parameter_B=2.5, param_type='Beta')
PDF = X.getPDF(300, graph=1)
```

<h2> Plot a Weibull distribution with shape parameters $\lambda=1$ and $k=5$

```python
X = Parameter(points=5, shape_parameter_A=1.0, shape_parameter_B=5.0, param_type='Weibull')
PDF = X.getSamples(10000, graph=1)
```
